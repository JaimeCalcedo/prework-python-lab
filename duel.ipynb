# Duel of sorcerers
## Intro
You are witnessing an epic battle between two powerful sorcerers: Gandalf and Saruman. Each sorcerer has 10 spells of variable power in their mind and they are going to throw them one after the other. The winner of the duel will be the one who wins more of those clashes between spells. Spells are represented as a list of 10 integers whose value equals the power of the spell.
```
gandalf = [10, 11, 13, 30, 22, 11, 10, 33, 22, 22]
saruman = [23, 66, 12, 43, 12, 10, 44, 23, 12, 17]
```
For example:
1. The first clash is won by Saruman: 10 against 23, wins 23
2. The second clash wins Saruman: 11 against 66, wins 66
3. etc.


You will create two variables, one for each sorcerer, where the sum of clashes won will be stored. Depending on which variable is greater at the end of the duel, you will show one of the following three results on the screen:
* Gandalf wins
* Saruman wins
* Tie

## Solution

````python
# Assign spell power lists to variables
gandalf = [10, 11, 13, 30, 22, 11, 10, 33, 22, 22]
saruman = [23, 66, 12, 43, 12, 10, 44, 23, 12, 17]
````

````python
# Assign 0 to each variable that stores the victories
gandalf_wins = 0
saruman_wins = 0
ties = 0
````

````python
# Execution of spell clashes
gandalf_wins = 0
saruman_wins = 0
ties = 0
result = zip()
result = zip(gandalf, saruman)
result_list = list(result)

if result_list [0][0] > result_list[0][1]: 
    gandalf_wins += 1
elif result_list [0][0] < result_list[0][1]:
    saruman_wins += 1
else: 
    ties += 1

if result_list [1][0] > result_list[1][1]: 
    gandalf_wins += 1
elif result_list [1][0] < result_list[1][1]:
    saruman_wins += 1
else: 
    ties += 1
    
if result_list [2][0] > result_list[2][1]: 
    gandalf_wins += 1
elif result_list [2][0] < result_list[2][1]:
    saruman_wins += 1
else: 
    ties += 1
    
if result_list [3][0] > result_list[3][1]: 
    gandalf_wins += 1
elif result_list [3][0] < result_list[3][1]:
    saruman_wins += 1
else: 
    ties += 1
    
if result_list [4][0] > result_list[4][1]: 
    gandalf_wins += 1
elif result_list [4][0] < result_list[0][1]:
    saruman_wins += 1
else: 
    ties += 1
    
if result_list [5][0] > result_list[5][1]: 
    gandalf_wins += 1
elif result_list [5][0] < result_list[5][1]:
    saruman_wins += 1
else: 
    ties += 1
    
if result_list [6][0] > result_list[6][1]: 
    gandalf_wins += 1
elif result_list [6][0] < result_list[6][1]:
    saruman_wins += 1
else: 
    ties += 1

if result_list [7][0] > result_list[7][1]: 
    gandalf_wins += 1
elif result_list [7][0] < result_list[7][1]:
    saruman_wins += 1
else: 
    ties += 1
    
if result_list [8][0] > result_list[8][1]: 
    gandalf_wins += 1
elif result_list [8][0] < result_list[8][1]:
    saruman_wins += 1
else: 
    ties += 1

if result_list [9][0] > result_list[9][1]: 
    gandalf_wins += 1
elif result_list [9][0] < result_list[9][1]:
    saruman_wins += 1
else: 
    ties += 1
    
print(gandalf_wins)
print(saruman_wins)
````

**Expected output**:
```

4
```

````python
# We check who has won, do not forget the possibility of a draw.
# Print the result based on the winner.
if gandalf_wins > saruman_wins: 
    print("Gandalf wins")
elif gandalf_wins < saruman_wins:
    print("Saruman wins")
else: 
    print("ItÂ´s a tie")
````
